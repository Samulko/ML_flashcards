```python
import numpy as np
import matplotlib.pyplot as plt

# --- Dataset ---
X = np.array([
    [1.0, 1.0],
    [1.5, 2],
    [3, 4],
    [2.5, 2.8],
    [4, 3.5],
    [5, 4]
])
y = np.array([1, 1, 1, -1, -1, -1])  # overlap + noise

# --- Generate all decision stumps ---
def generate_stumps(X, y):
    stumps = []
    for dim in range(X.shape[1]):
        sorted_vals = np.sort(np.unique(X[:, dim]))
        thresholds = [(sorted_vals[i] + sorted_vals[i + 1]) / 2 for i in range(len(sorted_vals) - 1)]
        for thresh in thresholds:
            for polarity in [+1, -1]:
                stumps.append({'dim': dim, 'thresh': thresh, 'polarity': polarity})
    return stumps

# --- Predict using a stump ---
def predict(stump, X):
    dim, thresh, polarity = stump['dim'], stump['thresh'], stump['polarity']
    pred = np.ones(len(X))
    if polarity == 1:
        pred[X[:, dim] > thresh] = -1
    else:
        pred[X[:, dim] <= thresh] = -1
    return pred

# --- Plot data points with weights and optional stump ---
def plot_data(X, y, w, stump=None, title=""):
    plt.figure(figsize=(5, 5))
    for i, (x, label) in enumerate(zip(X, y)):
        color = 'red' if label == 1 else 'blue'
        plt.scatter(*x, color=color, edgecolors='black', s=w[i] * 1000, marker='o')

    if stump:
        dim, thresh, polarity = stump['dim'], stump['thresh'], stump['polarity']
        if dim == 0:
            plt.axvline(thresh, color='black', linestyle='--')
            y_pos = plt.ylim()[0] + 0.1
            dx = 0.5 if polarity == 1 else -0.5
            plt.arrow(thresh, y_pos, dx, 0, head_width=0.1, head_length=0.2, fc='k', ec='k')
        else:
            plt.axhline(thresh, color='black', linestyle='--')
            x_pos = plt.xlim()[0] + 0.1
            dy = 0.5 if polarity == 1 else -0.5
            plt.arrow(x_pos, thresh, 0, dy, head_width=0.1, head_length=0.2, fc='k', ec='k')

    plt.title(title)
    plt.xlim(0.5, 5.5)
    plt.ylim(0.5, 4.5)
    plt.xlabel("$x_1$")
    plt.ylabel("$x_2$")
    plt.grid(True)
    plt.gca().set_aspect('equal')
    plt.tight_layout()
    plt.show()

# --- Plot decision surface ---
def plot_final_classifier(ensemble, X, y):
    x_min, x_max = 0.5, 5.5
    y_min, y_max = 0.5, 4.5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                         np.linspace(y_min, y_max, 300))
    grid = np.c_[xx.ravel(), yy.ravel()]
    f_vals = np.zeros(len(grid))
    for stump in ensemble:
        f_vals += stump['alpha'] * predict(stump, grid)
    f_vals = f_vals.reshape(xx.shape)

    plt.figure(figsize=(6, 5))
    im = plt.contourf(xx, yy, f_vals, levels=50, cmap='bwr', alpha=0.9)
    cbar = plt.colorbar(im)
    cbar.set_label('Classifier Output $f(x)$')

    for xi, label in zip(X, y):
        color = 'red' if label == 1 else 'blue'
        plt.plot(*xi, 'o', markersize=8, color=color, markeredgecolor='black')

    plt.xlabel("$x_1$")
    plt.ylabel("$x_2$")
    plt.title("Final Classifier $f$")
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.grid(False)
    plt.gca().set_aspect('equal')
    plt.tight_layout()
    plt.show()



# --- AdaBoost for a single path ---
def run_adaboost_path(X, y, first_stump, max_iter=5):
    n = len(y)
    w = np.ones(n) / n
    ensemble = []

    # --- First iteration ---
    stump = first_stump.copy()
    pred = predict(stump, X)
    err = np.sum(w * (pred != y))
    alpha = 0.5 * np.log((1 - err) / max(err, 1e-10))
    stump['alpha'] = alpha
    ensemble.append(stump)

    w *= np.exp(-alpha * y * pred)
    w /= np.sum(w)

    plot_data(X, y, w, stump, title=f"Start with stump: {stump} (α={alpha:.2f})")
    print(f"\nWeights after iteration 1:")
    for i in range(len(X)):
        print(f"  Point {i+1}: (x₁={X[i,0]}, x₂={X[i,1]}) → weight = {w[i]:.4f}")

    # --- Remaining iterations ---
    for m in range(1, max_iter):
        stumps = generate_stumps(X, y)
        best_err = float('inf')
        best_stump = None

        for stump in stumps:
            pred = predict(stump, X)
            err = np.sum(w * (pred != y))
            if err < best_err:
                best_err = err
                best_stump = stump.copy()

        alpha = 0.5 * np.log((1 - best_err) / max(best_err, 1e-10))
        best_stump['alpha'] = alpha
        ensemble.append(best_stump)

        pred = predict(best_stump, X)
        w *= np.exp(-alpha * y * pred)
        w /= np.sum(w)

        plot_data(X, y, w, best_stump, title=f"Iter {m+1}: {best_stump}, α={alpha:.2f}")
        print(f"\nWeights after iteration {m+1}:")
        for i in range(len(X)):
            print(f"  Point {i+1}: (x₁={X[i,0]}, x₂={X[i,1]}) → weight = {w[i]:.4f}")

        final_pred = np.sign(sum(s['alpha'] * predict(s, X) for s in ensemble))
        if np.all(final_pred == y):
            print(f"Path with stump {first_stump} stopped early at iteration {m+1}.")
            break

    return ensemble



# === Get all best stumps at iteration 1 ===
def get_initial_best_stumps(X, y):
    w = np.ones(len(y)) / len(y)
    stumps = generate_stumps(X, y)
    best_err = float('inf')
    best_stumps = []

    for stump in stumps:
        pred = predict(stump, X)
        err = np.sum(w * (pred != y))
        if err < best_err:
            best_err = err
            best_stumps = [stump.copy()]
        elif np.isclose(err, best_err):
            best_stumps.append(stump.copy())

    print("\n=== Initial Best Stumps ===")
    for s in best_stumps:
        print(s)
    return best_stumps

# === Run and Visualize for All Paths ===
initial_best = get_initial_best_stumps(X, y)

for i, stump in enumerate(initial_best):
    print(f"\n--- Running AdaBoost for Best Stump #{i+1} ---")
    ensemble = run_adaboost_path(X, y, stump, max_iter=5)
    plot_final_classifier(ensemble, X, y)
```
